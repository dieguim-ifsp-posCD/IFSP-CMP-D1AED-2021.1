<h1>Análise Exploratória de Dados</h1><hr/>

A **Análise Exploratória de Dados** refere-se ao processo de investigação de dados (de um problema) para a descoberta de _insights_, padrões, anomalias, testar hipóteses e checar suposições com a ajuda de técnicas e conceitos estatísticos e representações visuais/gráficas.

<h2>1. Dataset</h2>
<hr/>

**Dataset**: Gas Prices in Brazil: https://www.kaggle.com/matheusfreitag/gas-prices-in-brazil <br/>

Este dataset contém os **registros dos preços médios semanais dos combustíveis do Brasil entre os anos de 2004 e 2019**. <br/>

Alguns dos principais *atributos* (colunas) do dataset são: 'ESTADO', 'PRODUTO', 'NÚMERO DE POSTOS PESQUISADOS', 'PREÇO MÉDIO REVENDA'.

### 2.1 Como os preços da Gasolina Comum em São Paulo variaram em 2018?

```python
df = data.query('ESTADO == "SAO PAULO" and ANO == 2018 and PRODUTO == "GASOLINA COMUM"').reset_index(drop=True)
```

```python
df.head()
```

#### **Estatísticas Descritivas**

```python
df.describe()
```

```python
df['PREÇO MÉDIO REVENDA'].describe()
```

```python
df['PREÇO MÉDIO REVENDA'].describe().to_frame()
```

#### **Tendência Central dos Preços**

```python
print(f'O preço médio da Gasolina no Estado de São em 2018 foi de R$ {df["PREÇO MÉDIO REVENDA"].mean():.2f} ± {df["PREÇO MÉDIO REVENDA"].std():.2f}')
```

```python
plt.figure(figsize=(6,6))
ax = sns.barplot(y='PREÇO MÉDIO REVENDA', ci='sd', data=df)
ax.set_title('Gasolina Comum no Estado de SP em 2018')
ax.set_ylabel('Preço Médio (R$)')
```

```python
# "Truque" para colocar o valor da variável em cima da barra
plt.figure(figsize=(6,6))
ax = sns.barplot(y='PREÇO MÉDIO REVENDA', ci='sd', data=df)
ax.set_title('Gasolina Comum no Estado de SP em 2018')
ax.set_ylabel('Preço Médio (R$)')

for bar in ax.patches:
    height = bar.get_height()  # equivalent to counting
    x = bar.get_x() + bar.get_width()
    y = height
    ax.annotate(f'R$ {height:.2f}',
                xy=(x, y),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
```

```python
print(f'A mediana do preço da Gasolina Comum no estado de SP em 2018 é R$ {df["PREÇO MÉDIO REVENDA"].median():.2f}')
```

### **Variabilidade**

#### Histograma com número de bins padrão:

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(x='PREÇO MÉDIO REVENDA', data=df)
ax.set_title('Histograma dos Preços da Gasolina do Estado de SP em 2018')
ax.set_xlabel('Preço da Gasolina (R$)')
ax.set_ylabel('Frequência')
```
